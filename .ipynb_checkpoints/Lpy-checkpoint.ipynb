```python
from Core.trig import sin_p, cos_p, sin_chebyshev, cos_chebyshev
from Core.visual import plot_function, display_pi_of_p, display_transformation, ASVECTORFIELD, ASFUNCTION, ASVECTORSONCIRCLE
from Core.numeric import approx_pi
import matplotlib.pyplot as plt
from math import pi
```

```python
#Sinus und Cosinus so wie man es kennt: 
plot_function(lambda x: sin_p(x, 2), 0 , 2 * pi)
```

```python
plot_function(lambda x: cos_p(x, 2), 0 , 2 * pi)
```

```python
#Lpy bietet eine Generalisierte Version von Sinus:
#Leider unterstützt Lpy im Moment nur Metriken bis ungefähr l^16
plot_function(lambda x: sin_p(x, 1), 0 , 2 * pi)
```

```python
#Lpy bietet eine Generalisierte Version von Kosinus:
#Leider unterstützt Lpy im Moment nur Metriken bis ungefähr l^16
plot_function(lambda x: cos_p(x, 1), 0 , 2 * pi)
```

```python
#Lpy unterstützt auch die Chebyshev-Metrik
plot_function(sin_chebyshev, 0 , 2 * pi)
```

```python
#Lpy unterstützt auch die Chebyshev-Metrik
plot_function(cos_chebyshev, 0 , 2 * pi)
```

```python
#Auch kann Lpy den Wert von Pi nach Metrik bestimmen
print(approx_pi(5))
display_pi_of_p(1, 16)
```

Lpy kann auch wie in der PDF beschrieben die Unterschiede zwischen Metriken bzw. den Unterschied von einer Metrik zu l^2 darstellen. Hierführ gibt es drei Möglichkeiten:

```python
#Darstellen der Verhältnisfunktion
display_transformation(2, 1, mode=ASFUNCTION)
```

```python
#Darstellen eines Vektorfeldes, welches die Transformation von l^p1 zu l^p2 beschreibt.
display_transformation(2, 1, mode=ASVECTORFIELD)
```

```python
#Zeigt mit Vektoren wie sich der Einheitskreis verändert.
display_transformation(2, 1, mode=ASVECTORSONCIRCLE)
```
